```python
import pandas as pd
```

```python
#Toma datos del archivo .csv y los almacena en el dataframe
dataframe = pd.read_csv("Datos/house-price.csv", index_col=0)
#Tomamos el valor objetivo (variable objetivo)
y = dataframe['Price']
#Tomamos las caracteristicas (variables explicativas)
X = dataframe.drop(columns=['Price'])
```

```python
#En la libreria sklearn estan todos los algoritmos de machineLearning
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
#Computamos la ecuacion matematica
model.fit(X, y)
```

```python
#Calculamos las predicciones
model.predict(X)
```

```python
#Calcular la comparacion entre las predicciones y la realidad para ver que tan buena es la ecuacion matematica
model.score(X, y)
```

```python
CasaPrueba = dataframe.head(1)
CasaPrueba = CasaPrueba.drop(columns=['Price'])
#Predecimos el valor de una casa
model.predict(CasaPrueba)
```

```python
#Ver el arbol de decision que se utilizo
from sklearn.tree import plot_tree
plot_tree(model);
```

```python
import pickle
#Exportar modelo capaz de calcular el valor de las casas a una ruta para ser usado en una app.
with open("artifacts/model.pkl","wb") as f:
    pickle.dump(model,f)

#Importar modelo para verificarlo
with open("artifacts/model.pkl","rb") as ff:
    model2 = pickle.load(ff)

model2.predict(CasaPrueba)
```
